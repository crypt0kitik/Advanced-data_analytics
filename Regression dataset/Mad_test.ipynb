```python
import numpy as np
import pandas as pd

df = pd.read_csv("Numeric data (last year work).csv")

variable = "Rating"

# Calculate the median
median = df[variable].median()

# Calculate MAD
mad = np.median(np.abs(df[variable] - median))

# Define a threshold for noise detection
threshold = 3 * mad

# Detect noise
noise = df[np.abs(df[variable] - median) > threshold]

print("Detected noise using MAD:")
noise
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv("Numeric data (last year work).csv")

variable = "Year"

# Calculate the median
median = df[variable].median()

# Calculate MAD
mad = np.median(np.abs(df[variable] - median))

# Define a threshold for noise detection
threshold = 3 * mad

# Detect noise
noise = df[np.abs(df[variable] - median) > threshold]

print("Detected noise using MAD:")
noise
```
