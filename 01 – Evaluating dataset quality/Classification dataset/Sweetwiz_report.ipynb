```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics

import sweetviz as sv
```

```python
# upload data
df = pd.read_csv("bank.csv")
df.head()
```

```python
# analyze and create HTML-report
report = sv.analyze(df)
report.show_html('sweetviz_report_bank.html')
```

```python
# Insights:

# It’s easy to make a basic analysis about the number of rows, NaN values, 
# etc., because I don’t need to write many command lines to see this information.

# I will not go through basic info for every variable.
# I will focus on associations and relationships.

# Key Relationships:
# There is a notable positive correlation between duration and deposit. 
# This suggests that the longer a client is engaged during a contact, 
# the higher the likelihood of them making a deposit. 
# This is one of the strongest relationships in the matrix.
# But considering the analysis from the LIME and Phik matrix,
# it works the opposite: less time spent results in a more positive outcome.

# Poutcome (the outcome of the previous marketing campaign) also 
# has a strong correlation with deposit. 
# If the result of the previous campaign was successful, 
# it significantly influences the likelihood 
# of a deposit being made in the current campaign 
# because they trust the company and know how the system works.

# Pdays (the number of days since the client was last contacted) shows 
# a moderate positive correlation with deposit. 
# Clients who haven’t been contacted for a longer period 
# may be more likely to respond positively and make a deposit.
# As I mentioned earlier, it might be due to the annoyance of continuous ads.

# Good insight:
# Other demographic factors like age, education, and job 
# don’t show strong associations with the deposit outcome, 
# indicating that behavioral aspects (such as response to 
# previous campaigns or contact duration) are more critical 
# in this dataset.
```
